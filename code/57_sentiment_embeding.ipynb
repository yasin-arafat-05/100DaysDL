```python

import numpy as np 
import tensorflow as tf 
from tensorflow import keras
from keras import Sequential
from keras.utils import pad_sequences
from keras.activations import sigmoid
from keras.layers import SimpleRNN,Embedding,Dense
```

```python
docs = [ 'go india',
'india india',
'hip hip hurray',
'jeetega bhai jeetega india jeetega',
'bharat mata ki jai',
'kohli kohli',
'sachin sachin',
'dhoni dhoni',
'modi ji ki jai',
'inquilab zindabad'
]
```

```python
tokenize = tf.keras.preprocessing.text.Tokenizer()
tokenize.fit_on_texts(docs)
tokenize.index_word
```

```python
sequesces = tokenize.texts_to_sequences(docs)
sequesces
```

```python
sequesces = pad_sequences(sequesces,padding="post")
sequesces
```

```python

model = Sequential()

# Define your layers
model.add(Embedding(input_dim=18, output_dim=2))

# Print model summary
model.summary()
```

```python
model.compile('adam','accuracy')
```
